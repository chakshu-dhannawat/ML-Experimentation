```python
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt
```

```python
model_id = 'CompVis/ldm-text2im-large-256'
```

```python
ldm = DiffusionPipeline.from_pretrained(model_id)
```

```python
prompt = "beautiful sunset"
```

```python
images = ldm([prompt], num_inference_steps=200, eta=0.5, guidance_scale = 6)["sample"] 
```

```python
plt.imshow(images[0])
```

```python
prompt = " formula one race"
images = ldm([prompt], num_inference_steps=300, eta=0.5, guidance_scale = 6)["sample"] 
plt.imshow(images[0])
```
